```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\tryhardsinbioinfo\coursera ml\diabetes.csv")
print(df.head)
```

```python
df.dropna(inplace=True)
```

```python
x= df.drop('Outcome', axis=1).values #drop column named outcome along the columns axis=1 columwise
y=df['Outcome'].values.reshape(-1,1) #picks just the column outcome viz your target variable and .values converts to numpyarray .resahpe changes it from 1d to 2d
print(x.shape, y.shape)
```

```python
x=(x-x.mean(axis=0))/x.std(axis=0) #normalization of features so that each colum has mean 0 and std 1
```

```python
x_b = np.c_[np.ones((len(x), 1)), x] #add bias
```

```python
def sigmoid(z):  #adding sigmoid function
    return 1/(1+np.exp(-z))
```

```python
def compute_loss(y, y_pred):     #computing loss function
    m=len(y)
    epsilon= 1e-8
    return -(1/m) * np.sum(y*np.log(y_pred+epsilon)+ (1-y)*np.log(1-y_pred+epsilon))
```

```python
def gradient_descent(x,y, lr=0.01, epochs=1000):
    m,n= x.shape
    theta=np.zeros((n,1))
    losses=[]

    for _ in range(epochs):
        z=np.dot(x, theta)
        y_pred= sigmoid(z)
        loss=compute_loss(y, y_pred)
        gradient= np.dot(x.T, (y_pred-y))/m
        theta-= lr*gradient
        losses.append(loss)
    return theta, losses
 
```

```python
theta, losses = gradient_descent(x_b, y, lr=0.1, epochs=2000)
```

```python
y_pred = sigmoid(np.dot(x_b, theta))
y_pred_labels = (y_pred >= 0.5).astype(int)
accuracy = (y_pred_labels == y).mean() * 100
print(f"Training Accuracy: {accuracy:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```
