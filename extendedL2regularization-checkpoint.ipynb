```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\DELL\OneDrive\Documents\tryhardsinbioinfo\coursera ml\diabetes.csv")
print(df.head)               #load data
```

```python
df.dropna(inplace=True)  #clean and preprocess 
x=df.drop('Outcome', axis=1).values
y=df['Outcome'].values.reshape(-1,1)
```

```python
x=(x-x.mean(axis=0))/ x.std(axis=0) #normalize
```

```python
x_b= np.c_[np.ones((len(x), 1)), x]
```

```python
# Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Loss function with regularization
def compute_loss(x,y_pred, theta, lamda):
    m= len(y)
    base_loss=-(1/m)*np.sum(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))
    reg_term=(lamda/(2*m))*np.sum(np.square(theta[1:])) #exclude bais from penality 
    return base_loss + reg_term
```

```python
# Gradient descent with L2 regularization
def gradient_descent(x, y, lr=0.01, epochs=1000, lamda=0.1):
    m, n = x.shape
    theta = np.zeros((n, 1))
    losses = []

    for _ in range(epochs):
        z = np.dot(x, theta)
        y_pred = sigmoid(z)

        # Compute loss (with regularization)
        loss = compute_loss(y, y_pred, theta, lamda)
        
        # Regularized gradient
        gradient = (1 / m) * np.dot(x.T, (y_pred - y))
        gradient[1:] += (lamda / m) * theta[1:]  # Skip bias term for regularization

        # Update weights
        theta -= lr * gradient
        losses.append(loss)

    return theta, losses
```

```python
#train model
theta, losses= gradient_descent(x_b,y , lr=0.01, epochs=1500, lamda=0.5)
```

```python
#prediction
y_pred = sigmoid(np.dot(x_b, theta))
pred_class = (y_pred >= 0.5).astype(int)
```

```python
# Accuracy
accuracy = np.mean(pred_class == y)
```

```python
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss with L2 Regularization')
plt.show()
```
